# Guess-Who game with Human Anatomy

## Project Description

The goal of our project is to create a spin-off of a fun-interactive Guess Who game using organs in Human Biology for students. Being biology students ourselves who are exploring the possibilities of Python, the motivation behind our project is to enhance engagement and understanding of human biology concepts among students. Traditional methods of teaching can sometimes fail to captivate learners, especially since biology is a very dense and memorization heavy subject. Therefore, by making a spin-off of the popular Guess Who game that utilizes organs in human biology, we hope to use the inherent appeal of an interactive game to create fun learning experiences, or serve as a useful study tool for students that are looking for better ways to review important course content. This approach encourages students to think critically and immerse themselves in understanding different organs and concepts, while having fun. Overall through this approach, we strive to not only create an enjoyable game but also foster a better understanding of human biology by facilitating better learning.
The project involves creating functions and a dictionary to store names, images, and hints about organs and creates a user-friendly interface for students to learn about human anatomy. Users get three chances to guess each organ until they exhaust the total number of organs in the dictionary. Users can also choose to enter "quit" if they decide to stop playing the game. The scores for every user that plays are stored externally for reference. At the end of the game, the user is told how well they did in comparison to their own past scores and the scores of other users, and they can replay the game as many times as they'd like in a sitting to keep gaining new scores or come back to the game later with the same username.

### Project Code *continued...*

```python
# Starts the game and initializes variables
username, high_score, score_data = logic.start_game()

# Variable to control replaying the game
replay = True

# Loop to keep playing as long as the player wants
while replay == True:

    # Plays the organ guessing game and gets the score
    score = logic.play_organ_guessing_game()

    # Gives feedback on the user's score and their high score
    high_score = logic.check_user_scores(username, score, high_score, score_data)

    # Checks if the player wants to replay
    replay = logic.check_replay(replay)

    # Updates the scoresheet data
    score_data = logic.update_scoresheet(username, high_score, score_data)

    # Checks and displays high scores
    logic.check_high_scores(username, score_data)

# Removing test value from the dataframe if still present
if 'Test1' in score_data.index:
    score_data.drop(['Test1'], axis = 0, inplace = True)

# Saves the updated scoresheet to a CSV file
score_data.to_csv('scoresheet.csv')

# End message after the game finishes
print('Thanks for playing! Have a nice day!')
```
